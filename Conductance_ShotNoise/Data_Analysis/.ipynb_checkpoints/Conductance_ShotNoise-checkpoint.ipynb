```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_G_O = pd.read_csv(r'G_O.txt', sep='\t', header = None)
df_P_O = pd.read_csv(r'P_O.txt', sep='\t', header = None)

df_G_U = pd.read_csv(r'G_U.txt', sep='\t', header = None)
df_P_U = pd.read_csv(r'P_U.txt', sep='\t', header = None)

df_G_S = pd.read_csv(r'G_S.txt', sep='\t', header = None)
df_P_S = pd.read_csv(r'P_S.txt', sep='\t', header = None)
```

```python
df_G_S.rename(columns = lambda s: s+1, index = lambda s: s+1)
#The rows and columns represent the steps (billiards setups of chaotic ensemble) and the open channels, respectively.
```

```python
G_O = df_G_O.values
P_O = df_P_O.values

G_U = df_G_U.values
P_U = df_P_U.values

G_S = (df_G_S.values)/2
P_S = (df_P_S.values)/2
```

```python
fig, axs = plt.subplots(3,4, figsize = (12,8), constrained_layout = True)

fig.suptitle(r'DISTRIBUTION OF CONDUCTANCE FOR DIFFERENTS NUMBERS OF OPEN CHANNELS (N)', fontsize = 20)

for j in range(0,4):
    for i in range(0,3):
        
        axs[i,j].tick_params(axis='both', which = 'major', labelsize = 13)
        
        if (i == 0):
            axs[i,j].hist(G_O[:,j], bins = 40, color='dodgerblue', label = r'$\beta = 1$', alpha=1, density = True, histtype='bar', ec='black');
        else:
            if (i == 1):
                axs[i,j].hist(G_U[:,j], bins = 40, color='red', label = r'$\beta = 2$', alpha=1, density = True, histtype='bar', ec='black');
            else:
                if (i == 2):
                    axs[i,j].hist(G_S[:,j], bins = 40, color='lime', label = r'$\beta = 4$', alpha=1, density = True, histtype='bar', ec='black');
        
        if (j == 0):
            axs[0,j].set_title('N = 1', fontsize = 18)
            axs[i,0].set_ylabel(r'$\mathcal{P}_{\beta}(G)$', fontsize = 18)
            axs[i,0].legend(fontsize = 18)
        else:
            if (j == 3): 
                axs[0,3].set_title('N = 10', fontsize = 18)          
            else:
                axs[0,j].set_title('N = '+ str(j), fontsize = 18)
            
```

```python
fig, axs = plt.subplots(3,4, figsize = (12,8), constrained_layout = True)

fig.suptitle(r'DISTRIBUTION OF POWER SHOT NOISE FOR DIFFERENTS NUMBERS OF OPEN CHANNELS (N)', fontsize = 20)

for j in range(0,4):
    for i in range(0,3):
        
        axs[i,j].tick_params(axis='both', which = 'major', labelsize = 13)
        
        if (i == 0):
            axs[i,j].hist(P_O[:,j], bins = 40, color='dodgerblue', label = r'$\beta = 1$', alpha=1, density = True, histtype='bar', ec='black');
        else:
            if (i == 1):
                axs[i,j].hist(P_U[:,j], bins = 40, color='red', label = r'$\beta = 2$', alpha=1, density = True, histtype='bar', ec='black');
            else:
                if (i == 2):
                    axs[i,j].hist(P_S[:,j], bins = 40, color='lime', label = r'$\beta = 4$', alpha=1, density = True, histtype='bar', ec='black');
        
        if (j == 0):
            axs[0,j].set_title('N = 1', fontsize = 18)
            axs[i,0].set_ylabel(r'$\mathcal{P}_{\beta}(P)$', fontsize = 18)
            axs[i,0].legend(fontsize = 18)
        else:
            if (j == 3): 
                axs[0,3].set_title('N = 10', fontsize = 18)          
            else:
                axs[0,j].set_title('N = '+ str(j), fontsize = 18)
            
```

```python
meanG_O = np.mean(G_O, axis = 0)
meanG_U = np.mean(G_U, axis = 0)
meanG_S = np.mean(G_S, axis = 0)

varG_O = np.var(G_O, axis = 0)
varG_U = np.var(G_U, axis = 0)
varG_S = np.var(G_S, axis = 0)

meanP_O = np.mean(P_O, axis = 0)
meanP_U = np.mean(P_U, axis = 0)
meanP_S = np.mean(P_S, axis = 0)

varP_O = np.var(P_O, axis = 0)
varP_U = np.var(P_U, axis = 0)
varP_S = np.var(P_S, axis = 0)
```

```python
fig, axs = plt.subplots(2,2, figsize = (15,12), constrained_layout = True)

fig.suptitle(r'MEAN AND VARIANCE OF CONDUCTANCE AND POWER SHOT NOISE AS FUNCTION OF NUMBERS OF OPEN CHANNELS (N)', fontsize = 20)

##### Mean of Conductance #####

for i in range(0,len(meanG_O)):
    axs[0,0].scatter((i+1), meanG_O[i], s = 50, color = 'dodgerblue', alpha = 1)
    axs[0,0].scatter((i+1), meanG_U[i], s = 50, color = 'red', alpha = 1)
    axs[0,0].scatter((i+1), meanG_S[i], s = 50, color = 'lime', alpha = 1)

axs[0,0].plot([i for i in range(1,11)], meanG_O, label = r'$\beta = 1$', color = 'dodgerblue', linestyle = '-')
axs[0,0].plot([i for i in range(1,11)], meanG_U, label = r'$\beta = 2$', color = 'red', linestyle = '-')
axs[0,0].plot([i for i in range(1,11)], meanG_S, label = r'$\beta = 4$', color = 'lime', linestyle = '-')

axs[0,0].grid(linestyle=':')
axs[0,0].legend(fontsize = 20)

axs[0,0].set_title('MEAN OF CONDUCTANCE', fontsize = 20)
axs[0,0].set_xlabel('N', fontsize = 20)
axs[0,0].set_ylabel(r'$<G>$', fontsize = 20)
axs[0,0].tick_params(axis = 'both', which = 'major', labelsize = 18)

###### Variance of Conductance ######

for i in range(0,len(varG_O)):
    axs[1,0].scatter((i+1), varG_O[i], s = 50, color = 'dodgerblue', alpha = 1)
    axs[1,0].scatter((i+1), varG_U[i], s = 50, color = 'red', alpha = 1)
    axs[1,0].scatter((i+1), varG_S[i], s = 50, color = 'lime', alpha = 1)

axs[1,0].plot([i for i in range(1,11)], varG_O, label = r'$\beta = 1$', color = 'dodgerblue', linestyle = '-')
axs[1,0].plot([i for i in range(1,11)], varG_U, label = r'$\beta = 2$', color = 'red', linestyle = '-')
axs[1,0].plot([i for i in range(1,11)], varG_S, label = r'$\beta = 4$', color = 'lime', linestyle = '-')

axs[1,0].grid(linestyle=':')

axs[1,0].set_title('VARIANCE OF CONDUCTANCE', fontsize = 20)
axs[1,0].set_xlabel('N', fontsize = 20)
axs[1,0].set_ylabel(r'$var(G)$', fontsize = 20)
axs[1,0].tick_params(axis = 'both', which = 'major', labelsize = 18)

###### Mean of Power Shot Noise

for i in range(0,len(meanP_O)):
    axs[0,1].scatter((i+1), meanP_O[i], s = 50, color = 'dodgerblue', alpha = 1)
    axs[0,1].scatter((i+1), meanP_U[i], s = 50, color = 'red', alpha = 1)
    axs[0,1].scatter((i+1), meanP_S[i], s = 50, color = 'lime', alpha = 1)

axs[0,1].plot([i for i in range(1,11)], meanP_O, label = r'$\beta = 1$', color = 'dodgerblue', linestyle = '-')
axs[0,1].plot([i for i in range(1,11)], meanP_U, label = r'$\beta = 2$', color = 'red', linestyle = '-')
axs[0,1].plot([i for i in range(1,11)], meanP_S, label = r'$\beta = 4$', color = 'lime', linestyle = '-')

axs[0,1].grid(linestyle=':')

axs[0,1].set_title('MEAN OF POWER SHOT NOISE', fontsize = 20)
axs[0,1].set_xlabel('N', fontsize = 20)
axs[0,1].set_ylabel(r'$<P>$', fontsize = 20)
axs[0,1].tick_params(axis = 'both', which = 'major', labelsize = 18)


###### Variance of Power Shot Noise ######

for i in range(0,len(varG_O)):
    axs[1,1].scatter((i+1), varP_O[i], s = 50, color = 'dodgerblue', alpha = 1)
    axs[1,1].scatter((i+1), varP_U[i], s = 50, color = 'red', alpha = 1)
    axs[1,1].scatter((i+1), varP_S[i], s = 50, color = 'lime', alpha = 1)

axs[1,1].plot([i for i in range(1,11)], varP_O, label = r'$\beta = 1$', color = 'dodgerblue', linestyle = '-')
axs[1,1].plot([i for i in range(1,11)], varP_U, label = r'$\beta = 2$', color = 'red', linestyle = '-')
axs[1,1].plot([i for i in range(1,11)], varP_S, label = r'$\beta = 4$', color = 'lime', linestyle = '-')

axs[1,1].grid(linestyle=':')

axs[1,1].set_title('VARIANCE OF POWER SHOT NOISE', fontsize = 20)
axs[1,1].set_xlabel('N', fontsize = 20)
axs[1,1].set_ylabel(r'$var(P)$', fontsize = 20)
axs[1,1].tick_params(axis = 'both', which = 'major', labelsize = 18)
```
