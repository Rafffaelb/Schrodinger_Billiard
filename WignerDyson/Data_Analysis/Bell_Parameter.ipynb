```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_Bell_Parameter_Ress_O = pd.read_csv(r'Bell_Parameter/Bell_Ress/Bell_Parameter_O_Ress.txt', sep = '\t', header = None)
df_Bell_Parameter_Ress_U = pd.read_csv(r'Bell_Parameter/Bell_Ress/Bell_Parameter_U_Ress.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Ress_O.rename(columns = lambda s: ((s*25)+50))
```

```python
Bell_Parameter_Ress_O = df_Bell_Parameter_Ress_O.values
Bell_Parameter_Ress_U = df_Bell_Parameter_Ress_U.values

print(df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count())
Violations_O = df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count().values
```

```python
print(df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count())
Violations_U = df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count().values
```

```python
print("MAX VALUES OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Ress_O.max())
print("\n")
print("MAX VALUES OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Ress_U.max())
```

```python
fig, axs = plt.subplots(1,2, figsize = (14, 6), constrained_layout = True)

fig.suptitle(r'HISTOGRAM OF $P(|\mathcal{S}_{CHSH}|)$ FOR SCHRÃ–DINGER BILLIARD WITH 300 RESONANCES', fontsize = 20)

axs[0].hist(Bell_Parameter_Ress_O[:,10], bins = 50, color = 'dodgerblue', label = r'TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')
axs[0].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[0].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[0].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[0].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[0].axvline(2, color = 'black')

axs[1].hist(Bell_Parameter_Ress_U[:,10], bins = 50, color = 'red', label = r'No TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')
axs[1].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[1].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[1].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[1].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[1].axvline(2, color = 'black')
```

```python
fig, axs = plt.subplots(1,1, figsize = (8, 6), constrained_layout = True, facecolor = 'w')


axs.scatter([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_O, label = 'TRS', s = 100, color = 'dodgerblue', alpha = 1)
axs.scatter([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_U, label = 'No TRS', s = 100, color = 'red', alpha = 1)
    
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_O, color = 'dodgerblue', linestyle = 'dashed')
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_U, color = 'red', linestyle = 'dashed')

axs.set_ylim((0,0.8))

fig.suptitle(r'VIOLATIONS IN TERMS OF RESONANCES FOR SCHRODINGER BILLIARD', fontsize = 20)
axs.set_xlabel('Resonances', fontsize = 15)
axs.set_ylabel('Violations (%)', fontsize = 19)
axs.tick_params(axis = 'both', which = 'major', labelsize = 15)
axs.grid(linestyle = ':')
axs.legend(fontsize = 15)
```

```python
bins = np.arange(0,1 + 0.05, 0.05)
df_O = df_Bell_Parameter_O_Fixed_Base.groupby(pd.cut(df_Bell_Parameter_O_Fixed_Base[r'$\mathcal{C}$'], bins)).max()
Concurrence_O = df_O[r'$\mathcal{C}$'].values
Bell_Parameter_O = df_O[r'$\mathcal{S}_{CHSH}$'].values
df_O.head()
```

```python
df_Bell_Parameter_U_Fixed_Base.columns = [r'$\mathcal{S}_{CHSH}$', r'$\mathcal{C}$']
df_Bell_Parameter_U_Fixed_Base.sort_values(by=[r'$\mathcal{C}$'], ascending = True, inplace = True)
df_Bell_Parameter_U_Fixed_Base.head()
```

```python
df_U = df_Bell_Parameter_U_Fixed_Base.groupby(pd.cut(df_Bell_Parameter_U_Fixed_Base[r'$\mathcal{C}$'], bins)).max()
Concurrence_U = df_U[r'$\mathcal{C}$'].values
Bell_Parameter_U = df_U[r'$\mathcal{S}_{CHSH}$'].values
df_U.head()
```

```python
fig, axs = plt.subplots(1,1, figsize = (8, 6), constrained_layout = True, facecolor = 'w')

axs.scatter(Concurrence_O, (1/2)*Bell_Parameter_O, label = r'Fixed Base ($\beta = 1$)', s = 50, color = 'dodgerblue', alpha = 1, marker = 'o')
axs.plot(Concurrence_O, (1/2)*Bell_Parameter_O, color = 'dodgerblue', linestyle = '-')

fig.suptitle(r'MAXIMAL VIOLATION OF BELL INEQUALITY AS FUNCTION OF CONCURRENCE ($\beta = 1$)', fontsize = 20)

axs.scatter(Concurrence_U, (1/2)*Bell_Parameter_U, label = r'Fixed Base ($\beta = 2$)', s = 50, color = 'red', alpha = 1, marker = 'o')
axs.scatter(Concurrence_U, np.sqrt(1+np.power(Concurrence_U,2)), label = 'Pure States', s = 50, color = 'orange', alpha = 1, marker = 's')
    
axs.plot(Concurrence_U, (1/2)*Bell_Parameter_U, color = 'red', linestyle = '-')
axs.plot(Concurrence_U, np.sqrt(1+np.power(Concurrence_U,2)), color = 'orange', linestyle = '-')


fig.suptitle(r'MAXIMAL VIOLATION AS FUNCTION OF CONCURRENCE FOR SCHRODINGER BILLIARD', fontsize = 20)

axs.set_ylim((1,1.45))
axs.set_xlabel(r'$\mathcal{C}$', fontsize = 19)
axs.set_ylabel(r'$\mathcal{S}_{CHSH, max}/2$', fontsize = 19)
axs.tick_params(axis = 'both', which = 'major', labelsize = 15)
axs.grid(linestyle = ':', linewidth = 2)
axs.legend(fontsize = 15)
    
```
