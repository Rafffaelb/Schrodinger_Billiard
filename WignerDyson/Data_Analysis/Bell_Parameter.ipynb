```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_Bell_Parameter_Ress_O = pd.read_csv(r'Bell_Parameter/Bell_Ress/Bell_Parameter_O_Ress.txt', sep = '\t', header = None)
df_Bell_Parameter_Ress_U = pd.read_csv(r'Bell_Parameter/Bell_Ress/Bell_Parameter_U_Ress.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Ress_O.rename(columns = lambda s: ((s*25)+50))
```

```python
df_Bell_Parameter_Ress_U.rename(columns = lambda s: ((s*25)+50))
```

```python
Bell_Parameter_Ress_O = df_Bell_Parameter_Ress_O.values
Bell_Parameter_Ress_U = df_Bell_Parameter_Ress_U.values

df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count()
Violations_O = df_Bell_Parameter_Ress_O[df_Bell_Parameter_Ress_O > 2.0].count().values
```

```python
df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count()
Violations_U = df_Bell_Parameter_Ress_U[df_Bell_Parameter_Ress_U > 2.0].count().values
```

```python
print("MAX VALUES OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Ress_O.max())
print("\n")
print("MAX VALUES OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Ress_U.max())
```

```python
fig, axs = plt.subplots(1,2, figsize = (14, 6), constrained_layout = True)

fig.suptitle(r'Histogram of $P(|\mathcal{S}_{CHSH}|)$ for SchrÃ¶dinger Billiard with 300 resonances', fontsize = 20)

axs[0].hist(Bell_Parameter_Ress_O[:,10], bins = 50, color = 'dodgerblue', label = r'TRS', alpha = 1, histtype = 'bar', ec = 'black')
axs[0].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[0].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[0].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[0].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[0].axvline(2, color = 'black')

axs[1].hist(Bell_Parameter_Ress_U[:,10], bins = 50, color = 'red', label = r'No TRS', alpha = 1, histtype = 'bar', ec = 'black')
axs[1].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
axs[1].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
axs[1].tick_params(axis = 'both', which = 'major', labelsize = 20)
axs[1].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
axs[1].axvline(2, color = 'black')
```

```python
for i in range(len(Violations_O)):
    plt.scatter(25*i + 50, (100/1000000)*Violations_O[i], s = 100, color = 'dodgerblue', alpha = 1)
    plt.scatter(25*i + 50, (100/1000000)*Violations_U[i], s = 100, color = 'red', alpha = 1)
    
plt.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_O, label = 'TRS', color = 'dodgerblue', linestyle = '-')
plt.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_U, label = 'No TRS', color = 'red', linestyle = '-')

plt.ylim((0,0.3))

plt.title(r'VIOLATIONS OF BELL INEQUALITY AS FUNCTION OF RESONANCES (N = 2, $\Gamma = 1$)', fontsize = 20)
plt.xlabel('Resonances', fontsize = 15)
plt.ylabel('Violations (%)', fontsize = 19)
plt.tick_params(axis = 'both', which = 'major', labelsize = 15)
plt.grid(linestyle = ':')
plt.legend(fontsize = 15)
plt.figure(figsize = (12,10))
```

```python
df_Bell_Parameter_Gamma_O = pd.read_csv(r'Bell_Parameter/Bell_Gamma/Bell_Parameter_O_Gamma.txt', sep = '\t', header = None)
df_Bell_Parameter_Gamma_U = pd.read_csv(r'Bell_Parameter/Bell_Gamma/Bell_Parameter_U_Gamma.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Gamma_O.rename(columns = lambda s: (s/20))
```

```python
df_Bell_Parameter_Gamma_U.rename(columns = lambda s: (s/20))
```

```python
print("MAX VALUES OF BELL PARAMETER IN ORTHOGONAL ENSEMBLE")
print(df_Bell_Parameter_Gamma_O.max())
print("\n")
print("MAX VALUES OF BELL PARAMETER IN UNITARY ENSEMBLE")
print(df_Bell_Parameter_Gamma_U.max())
```

```python
Violations_Gamma_O = df_Bell_Parameter_Gamma_O[df_Bell_Parameter_Gamma_O > 2.0].count().values
print(df_Bell_Parameter_Gamma_O[df_Bell_Parameter_Gamma_O > 2.0].count())
print("\n")
Violations_Gamma_U = df_Bell_Parameter_Gamma_U[df_Bell_Parameter_Gamma_O > 2.0].count().values
print(df_Bell_Parameter_Gamma_U[df_Bell_Parameter_Gamma_U > 2.0].count())
```

```python
for i in range(len(Violations_Gamma_O)):
    plt.scatter(i/20, (100/1000000)*Violations_Gamma_O[i], s = 150, color = 'dodgerblue', alpha = 1)
    plt.scatter(i/20, (100/1000000)*Violations_Gamma_U[i], s = 100, color = 'red', alpha = 1)
    
plt.plot([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_O, label = 'TRS', color = 'dodgerblue', linestyle = '-')
plt.plot([(i/20) for i in range(0,21)], (100/1000000)*Violations_Gamma_U, label = 'No TRS', color = 'red', linestyle = '-')

plt.ylim((0,0.3))

plt.title(r'VIOLATIONS OF BELL INEQUALITY AS FUNCTION OF $\Gamma$', fontsize = 20)
plt.xlabel(r'$\Gamma$', fontsize = 15)
plt.ylabel('Violations (%)', fontsize = 19)
plt.tick_params(axis = 'both', which = 'major', labelsize = 15)
plt.grid(linestyle = ':')
plt.legend(fontsize = 15)
plt.figure(figsize = (12,10))
```
